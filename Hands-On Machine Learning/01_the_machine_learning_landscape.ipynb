#### 실시간으로 점진적인 학습을 하는지 아닌지 (온라인 학습과 배치 학습)
- 배치학습 (오프라인 학습)
    - 가용한 데이터를 모두 사용하여 훈련
    - 학습한 것을 적용 (점진적인 학습 불가능)
    - 훈련에 많은 시간이 소요됨
        - 컴퓨팅 자원 확대
        - 데이터 양 축소
        - 자원이 제한된 시스템이 많은 양의 데이터를 활용하면 문제가 될 수도 있음
    - 온라인 학습
        - 데이터를 순차적으로 한 개씩 또는 미니배치라 부르는 작은 묶음 단위로 주입하여 시스템을 훈련
        - 적합한 상황
            - 연속적으로 데이터를 받고 빠른 변화에 스스로 적응해야 하는 시스템에 적합
            - 제한된 컴퓨팅 자원 환경에 적합
            - 지난 학습 데이터는 버려 메모리 절약
            - 대량의 데이터셋 학습에 적합 (외부 메모리 학습)
        - 학습률 : 변화하는 데이터에 얼마나 빠르게 적응할 것인지 나타냄
            - 학습률이 높을 경우 : 시스템이 데이터에 빠르게 적응하지만 이전 데이터를 빨리 잊음
            - 학습률이 낮을 경우 : 시스템의 관성이 더 커져 더 느리게 학습 되지만 새로운 데이터에 있는 잡음이나 대표성 없는 데이터 포인트에 덜 민감
        - 문제점
            - 시스템에 나쁜 데이터가 주입되었을 때 시스템 성능이 점진적으로 감소
            - 잦은 모니터링 및 성능 감소시 즉각 학습 중지
        
#### 단순하게 알고 있는 데이터 포인트와 새 데이터 포인트를 비교하는 것인지 아니면 과학자들이 하는 것처럼 훈련 데이터셋에서 패턴을 발견하여 예측 모델을 만드는지 (사례 기반 학습과 모델 기반 학습)
- 사례 기반 학습
    - 시스템이 훈련 샘플을 기억함으로써 학습한다.
    - 그 후 유사도 측정을 사용해 새로운 데이터와 학습한 샘플을 비교하는 것으로 일반화 한다.
- 모델 기반 학습
    - 샘플의 모델을 만들어 예측에 사용
    - 순서
        - 모델 선택
        - 효용 함수 또는 비용함수로 모델 파라미터 조정
        - 효용 함수 : 모델이 얼마나 좋은지 측정
        - 비용 함수 : 모델이 얼마나 나쁜지 측정

이 함수는 OECD의 삶의 만족도(life satisfaction) 데이터와 IMF의 1인당 GDP(GDP per capita) 데이터를 합칩니다. 이는 번거로운 작업이고 머신러닝과는 관계가 없기 때문에 책 안에 포함시키지 않았습니다.

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

#### 나쁜 데이터 사례
- 충분하지 않은 양의 훈련 데이터
    - 복잡한 문제에서 알고리즘보다는 데이터가 더 중요, 하지만 작거나 중간규모의 데이터셋이 흔하므로 알고리즘을 무시해서는 안됨
- 대표성 없는 훈련 데이터
    - 샘플이 작으면 샘플링 잡음 문제 발생
    - 표본 추출 방법이 잘못되면 샘플링 편향 문제 발생
- 낮은 품질의 데이터
    - 일부 샘플이 이상치라는 게 명확하면 그것들을 무시하거나 수동으로 잘못된 것을 고침
    - 일부 샘플에 특성 몇 개가 빠져있다면, 이 특성을 모두 무시할지, 이 샘플을 무시할지, 빠진 값을 채울지, 또는 이 특성을 넣은 모델과 제외한 모델을 따로 훈련시킬 것인지 결정해야함
- 관련 없는 특성
    - 특성 선택 : 가지고 있는 특성 중에서 훈련에 가장 유용한 특성을 선택
    - 특성 추출 : 특성을 결합하여 더 유용한 특성을 만듬. 차원 축소 알고리즘이 도움이 될 수 있음
    - 새로운 데이터 수집으로 새 특성 만듦

#### 나쁜 알고리즘 사례
- 훈련 데이터 과대적합
    - 훈련 데이터에 있는 잡음의 양에 비해 모델이 너무 복잡할 때 일어남
    - 하이퍼파라미터 : 학습 알고리즘의 파라미터
    - 해결 방법
        - 파라미터 수가 적은 모델 선택
        - 훈련 데이터에 있는 특성 수 줄임
        - 모델에 제약을 가하여 단순화
        - 훈련 데이터를 더 많이 모음
        - 훈련 데이터의 잡음을 줄임 (오류데이터 수정 및 이상치 제거)
- 훈련 데이터 과소적합
    - 모델이 너무 단순해서 데이터의 내재된 구조를 학습하지 못함
    - 해결 방법
        - 모델 파라미터가 더 많은 강력한 모델 선택
        - 학습 알고리즘에서 더 좋은 특성 제공 (특성 공학)
        - 모델의 제약을 줄임 (규제 하이퍼파라미터 감소)

### 하이퍼파라미터 튜닝 및 모델 선택
- 홀드아웃 검증 : 훈련 세트의 일부를 떼어내어 여러 후보 모델을 평가하고 가장 좋은 하나 선택
- 검증 세트를 사용하여 최적 하이터파라미터를 선택하여 최종 모델에 적용
- 교차 검증 : 작은 검증 세트를 여러 개를 사용

#### 테스트 세트에 사용되는 데이터와 훈련 세트 사이에 데이터 불일치 위험이 있을 때
- 훈련-개발 세트 : 훈련 데이터의 일부를 떼어내어 또 다른 세트를 만듬
- 모델을 훈련 세트에서 훈련한 후 훈련-개발 세트에서 평가
- 모델이 훈련 세트에서 잘 동작하지만 훈련-개발 세트에서 나쁜 성능을 낸다면 훈련 세트에 과대 적합 되었을 가능성이 높음
- 훈련 세트와 훈련-개발 세트에서 모두 잘 동작하지만 개발 세트에서 성능이 나쁘다면 훈련 데이터와 검증, 테스트 데이터 사이에 불일치가 있을 가능성이 높음
- 검증, 테스트 데이터에 더 가깝게 훈련 데이터를 개선해야 함
